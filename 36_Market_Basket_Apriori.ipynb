#  Analiza Koszykowa: Reguy Asocjacyjne (Apriori)

To technika u偶ywana przez ka偶dy supermarket i sklep e-commerce (np. "Czsto kupowane razem" na Amazonie).

Musimy zrozumie 3 metryki:

1.  **Support (Wsparcie):** Jak popularny jest zestaw?
    *   $Support(A) = \frac{\text{Transakcje z A}}{\text{Wszystkie transakcje}}$
    *   Niskie wsparcie = towar niszowy.
2.  **Confidence (Ufno):** Jeli kupiem A, to na ile % kupi B?
    *   $Confidence(A \to B) = \frac{\text{Transakcje z A i B}}{\text{Transakcje z A}}$
    *   To nasza "sia reguy".
3.  **Lift (Przyrost):** Czy ta zale偶no jest prawdziwa, czy to przypadek?
    *   $Lift = 1$: Brak zwizku (Niezale偶ne).
    *   $Lift > 1$: A przyciga B (Super!).
    *   $Lift < 1$: A odpycha B (Kto kupuje A, ten rzadziej kupuje B).

Algorytm **Apriori** przeszukuje baz transakcji i wypluwa reguy typu: `{Chleb} -> {Maso}`.

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# 1. DANE TRANSAKCYJNE (Lista list)
# Ka偶dy wiersz to jeden paragon.
dataset = [
    ['Mleko', 'Cebula', 'Gaka muszkatoowa', 'Jajka', 'Jogurt'],
    ['Koper', 'Cebula', 'Gaka muszkatoowa', 'Jajka', 'Jogurt'],
    ['Mleko', 'Jabko', 'Jajka'],
    ['Mleko', 'Jednoro偶ec', 'Kukurydza', 'Jajka', 'Jogurt'],
    ['Kukurydza', 'Cebula', 'Cebula', 'Jajka', 'Lody', 'Jajka']
]

print("--- PRZYKADOWY PARAGON ---")
print(dataset[0])
```

## Krok 1: One-Hot Encoding (Format 0/1)

Algorytm Apriori nie rozumie napis贸w. Rozumie tabel prawdy.
Musimy zamieni nasz list zakup贸w na wielk tabel, gdzie:
*   Kolumny = Wszystkie mo偶liwe produkty w sklepie.
*   Wiersze = Paragony.
*   Warto = True (kupi) / False (nie kupi).

U偶yjemy do tego `TransactionEncoder`.

```python
# Inicjalizacja Encodera
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)

# Zamiana na DataFrame (偶eby byo adnie wida)
df = pd.DataFrame(te_ary, columns=te.columns_)

print("--- TABELA PRAWDA/FASZ (One-Hot) ---")
display(df)
print(f"Mamy {df.shape[0]} transakcji i {df.shape[1]} unikalnych produkt贸w.")
```

## Krok 2: Szukanie "Czstych Zbior贸w" (Frequent Itemsets)

Teraz uruchamiamy **Apriori**.
M贸wimy mu: *"Poka偶 mi tylko te zestawy produkt贸w, kt贸re wystpuj w przynajmniej 60% paragon贸w (`min_support=0.6`)"*.

To odsieje rzadkie produkty (jak "Jednoro偶ec" czy "Lody"), a zostawi bestsellery.

```python
# use_colnames=True sprawia, 偶e widzimy nazwy produkt贸w, a nie numery kolumn
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)

# Sortujemy od najczstszych
frequent_itemsets = frequent_itemsets.sort_values(by="support", ascending=False)

print("--- NAJCZSTSZE ZESTAWY (Support > 60%) ---")
print(frequent_itemsets)

print("\nWniosek: Jajka s w 100% koszyk贸w (Support=1.0). Kr贸lowie sprzeda偶y.")
```

## Krok 3: Generowanie Regu (Association Rules)

Mamy popularne zestawy (np. Jajka i Cebula).
Ale czy to znaczy, 偶e **Jajka powoduj kupno Cebuli**? Czy na odwr贸t?
A mo偶e po prostu wszyscy kupuj jajka, wic si rzeczy lduj one z cebul?

Tu wchodzi metryka **Lift**.
Generujemy reguy, kt贸re maj np. minimum 70% ufnoci (`min_threshold=0.7`).

```python
# Tworzymy reguy na podstawie zbior贸w z kroku 2
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Wybieramy tylko interesujce kolumny
cols = ['antecedents', 'consequents', 'support', 'confidence', 'lift']
rules_clean = rules[cols].sort_values(by='lift', ascending=False)

print("--- WYKRYTE REGUY ASOCJACYJNE ---")
# antecedents = Jeli kupie TO...
# consequents = ...to kupisz TAMTO
display(rules_clean)
```

##  Podsumowanie: Dlaczego Lift jest najwa偶niejszy?

Sp贸jrz na wyniki.

1.  **Regua:** `{Cebula} -> {Jajka}`
    *   **Confidence:** 1.0 (100%). Ka偶dy kto kupi cebul, kupi jajka.
    *   **Lift:** 1.0.
    *   **Interpretacja:** Lift = 1 oznacza NUDA. Jajka s tak popularne (s wszdzie), 偶e kupno cebuli nie zwiksza szansy na kupno jajek. One i tak by tam byy. To nie jest "zale偶no", to statystyka.

2.  **Gdyby mia regu:** `{Piwo} -> {Chipsy}`
    *   Gdyby **Lift** wynosi **2.5**, oznaczaoby to, 偶e klient z piwem jest **2.5 raza bardziej skonny** kupi chipsy ni偶 losowy klient.
    *   **To jest Zoto.** To tutaj sklep stawia stojak z chipsami obok lod贸wki z piwem.

**Wniosek:**
W analizie koszykowej nie patrz lepo na Confidence (Ufno). Patrz na **Lift**.
Lift > 1 to pienidz.
Lift = 1 to przypadek.
