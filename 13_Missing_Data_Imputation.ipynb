# Ч Obsuga Brakujcych Danych (Missing Data)

W prawdziwym wiecie dane nigdy nie s czyste. Zawsze czego brakuje:
*   Klient nie poda wieku.
*   Awaria czujnika temperatury.
*   Bd zapisu w bazie.

W Pythonie braki oznaczamy jako **NaN** (Not a Number).
Wikszo modeli ML (Regresja, Sieci Neuronowe) **wywali si**, jeli zobaczy NaN.

Mamy trzy g贸wne strategie:
1.  **Usuwanie:** "Nie ma danych? Do mieci." (Ryzykowne).
2.  **Imputacja Prosta:** "Wstaw redni." (Szybkie, ale znieksztaca).
3.  **Imputacja Inteligentna:** "Zgadnij warto na podstawie grupy." (Najlepsze).

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# 1. TWORZYMY BRUDNE DANE (Symulacja HR)
# Mamy pracownik贸w. Niekt贸rzy nie podali WIEKU, inni PENSJI.
data = {
    'Stanowisko': ['Junior', 'Senior', 'Junior', 'Senior', 'Junior', 'Senior', 'Boss'],
    'Wiek':       [22,       45,       np.nan,   50,       23,       np.nan,   55],
    'Pensja':     [4000,     12000,    3500,     np.nan,   np.nan,   13000,    25000]
}

df = pd.DataFrame(data)

print("--- ORYGINA (Zauwa偶 NaN) ---")
display(df)

# Sprawd藕my, ile dokadnie brakuje
print("\nIle brak贸w w ka偶dej kolumnie?")
print(df.isnull().sum())
```

## Metoda 1: Usuwanie (The Nuclear Option)

Najprostsze podejcie: `dropna()`.
Usuwamy ka偶dy wiersz, kt贸ry ma chocia偶 jedn dziur.

```python
df_dropped = df.dropna()

print(f"Oryginalny rozmiar: {df.shape}")
print(f"Po usuniciu brak贸w: {df_dropped.shape}")
print("-" * 30)

print("--- WYNIK ---")
display(df_dropped)

print("\nWNIOSEK: Stracilimy prawie poow danych! (3 z 7 wierszy).")
print("Jeli masz mao danych, to samob贸jstwo.")
```

## Metoda 2: Prosta Imputacja (rednia)

Zamiast usuwa, wstawmy redni z caej kolumny.
U偶yjemy do tego `SimpleImputer` z biblioteki Scikit-Learn.

**Problem:**
rednia pensja w firmie to np. 10 000.
Jeli wstawimy t liczb Juniorowi (kt贸ry zarabia 4000), to sztucznie go "wzbogacimy". Model zgupieje.

```python
# Tworzymy Imputer (Strategia: rednia)
imputer = SimpleImputer(strategy='mean')

# Kopiujemy dane, 偶eby nie zepsu oryginau
df_mean = df.copy()

# Wybieramy tylko kolumny liczbowe (Stanowiska nie da si uredni matematycznie)
cols_num = ['Wiek', 'Pensja']
df_mean[cols_num] = imputer.fit_transform(df[cols_num])

print("--- IMPUTACJA REDNI ---")
display(df_mean)
print("Sp贸jrz na wiersz 4 (Junior). Dosta pensj ok. 11 500.")
print("To nie ma sensu! Juniorzy tyle nie zarabiaj.")
```

## Metoda 3: Imputacja Kontekstowa (Grupowanie)

To jest **sprytne podejcie**.
Zamiast bra redni wszystkich ludzi, we藕my:
*   Dla brakujcego wieku Juniora -> redni wieku **innych Junior贸w**.
*   Dla brakujcej pensji Seniora -> redni pensj **innych Senior贸w**.

U偶yjemy do tego magii Pandasa: `groupby() + transform()`.

```python
df_smart = df.copy()

# 1. Wypeniamy PENSJ redni w ramach STANOWISKA
df_smart['Pensja'] = df_smart['Pensja'].fillna(
    df_smart.groupby('Stanowisko')['Pensja'].transform('mean')
)

# 2. Wypeniamy WIEK redni w ramach STANOWISKA
df_smart['Wiek'] = df_smart['Wiek'].fillna(
    df_smart.groupby('Stanowisko')['Wiek'].transform('mean')
)

print("--- IMPUTACJA INTELIGENTNA (Grupowa) ---")
display(df_smart)

print("ANALIZA:")
print("- Junior (wiersz 4) dosta pensj 3750 (rednia Junior贸w). Bardzo realne!")
print("- Senior (wiersz 3) dosta pensj 12500 (rednia Senior贸w). Te偶 realne.")
```

##  Podsumowanie: Nie kam danymi

Dlaczego spos贸b atania dziur ma znaczenie?

**Tu jest haczyk.**
Ka偶da imputacja to troch **kamstwo**. Wymylamy dane, kt贸rych nie byo.
*   Jeli usuniesz dane -> Wprowadzasz **Bias** (np. usuwasz tylko biednych ludzi, bo oni rzadziej podaj zarobki, wic Tw贸j model myli, 偶e wiat jest bogaty).
*   Jeli wstawisz redni globaln -> Zabijasz r贸偶norodno (**Variance**). Wszyscy staj si "przecitni".
*   Jeli u偶yjesz kontekstu (Grupy) -> Twoje "kamstwo" jest najbardziej prawdopodobne.
