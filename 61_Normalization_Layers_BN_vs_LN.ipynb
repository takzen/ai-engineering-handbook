#  Normalization Layers: Batch vs Layer vs Instance

Bez normalizacji gbokie sieci nie dziaaj (gradienty wybuchaj lub znikaj).
Musimy sprowadzi dane do wsp贸lnego mianownika (rednia ~0, odchylenie ~1).

Ale jak to policzy, gdy mamy tensor 4D? `[Batch, Channel, Height, Width]`.

1.  **Batch Norm (BN):** Patrzy na **ten sam kana** we wszystkich zdjciach w batchu.
    *   *"Jaki jest redni kolor czerwony w caym batchu?"*
    *   *Kr贸l CNN.* Wymaga du偶ego Batch Size.
2.  **Layer Norm (LN):** Patrzy na **wszystkie kanay** w jednym zdjciu.
    *   *"Jaki jest redni sygna w tym konkretnym zdaniu?"*
    *   *Kr贸l NLP (Transformer贸w).* Dziaa nawet przy Batch Size = 1.
3.  **Instance Norm (IN):** Patrzy na **jeden kana w jednym zdjciu**.
    *   *Kr贸l Style Transfer (GAN).* Ignoruje kontrast zdjcia.

Zrobimy eksperyment na maym tensorze, 偶eby zobaczy r贸偶nic w liczbach.

```python
import torch
import torch.nn as nn

# 1. PRZYGOTOWANIE DANYCH
# Tensor [Batch=2, Channels=3, H=2, W=2]
# Wyobra藕 sobie 2 malutkie obrazki 2x2 piksele RGB.

# Obrazek 1: Ciemny (same mae liczby: 0, 1, 2)
img1 = torch.tensor([
    [[0., 1.], [0., 1.]], # R
    [[2., 3.], [2., 3.]], # G
    [[0., 0.], [0., 0.]]  # B
])

# Obrazek 2: Jasny (same du偶e liczby: 10, 11, 12)
img2 = torch.tensor([
    [[10., 11.], [10., 11.]], # R
    [[12., 13.], [12., 13.]], # G
    [[10., 10.], [10., 10.]]  # B
])

# Sklejamy w Batch
batch = torch.stack([img1, img2])

print(f"Ksztat: {batch.shape} (N, C, H, W)")
print("Obrazek 1: Mae wartoci (rednia ~1.5)")
print("Obrazek 2: Du偶e wartoci (rednia ~11.5)")
```

## 1. Batch Normalization (Pionowo)

BN normalizuje "wzdu偶 Batcha".
Dla Kanau R (Czerwonego) we藕mie piksele z Obrazka 1 **ORAZ** z Obrazka 2.

*   rednia Kanau R = (rednia img1 + rednia img2) / 2
*   Efekt: Obrazek 1 zostanie "podcignity" w g贸r, a Obrazek 2 "cignity" w d贸. **Tracimy informacj, 偶e jeden by ciemny, a drugi jasny!** Ale zyskujemy stabilno cech.

```python
# Inicjalizacja BatchNorm (num_features = liczba kana贸w = 3)
bn = nn.BatchNorm2d(num_features=3)

# Przepuszczamy dane (w trybie treningowym, 偶eby liczy statystyki)
bn.train()
out_bn = bn(batch)

print("--- BATCH NORM ---")
print("Obrazek 1 (Ciemny):\n", out_bn[0, 0, :, :]) # Kana R
print("Obrazek 2 (Jasny):\n", out_bn[1, 0, :, :]) # Kana R

print("\nWNIOSEK:")
print("Wartoci w obu obrazkach s teraz podobne (ok. -1 i +1).")
print("BN wymiesza statystyki obu obrazk贸w.")
```

## 2. Layer Normalization (Poziomo)

LN normalizuje "wewntrz Obrazka".
Dla Obrazka 1 we藕mie wszystkie jego piksele (R, G, B) i policzy ich redni.
Nie obchodzi go Obrazek 2.

*   Efekt: Zachowuje niezale偶no pr贸bek. Dlatego jest kluczowy w NLP, gdzie jedno zdanie mo偶e by dugie, a drugie kr贸tkie, i nie chcemy, 偶eby jedno wpywao na drugie.

```python
# Inicjalizacja LayerNorm
# Musimy poda ksztat tego, co normalizujemy (C, H, W)
ln = nn.LayerNorm([3, 2, 2])

out_ln = ln(batch)

print("--- LAYER NORM ---")
print("Obrazek 1:\n", out_ln[0, 0, :, :])
print("Obrazek 2:\n", out_ln[1, 0, :, :])

print("\nWNIOSEK:")
print("Ka偶dy obrazek zosta znormalizowany OSOBNO.")
```

## 3. Instance Normalization (Styl)

IN idzie o krok dalej. Normalizuje ka偶dy kana w ka偶dym obrazku osobno.
*   Obrazek 1, Kana R -> Osobna rednia.
*   Obrazek 1, Kana G -> Osobna rednia.

**Dlaczego to wa偶ne?**
To usuwa "styl" (kontrast, jasno globaln) z konkretnego koloru. Dlatego jest u偶ywane w **Style Transfer** (np. przerabianie zdjcia na obraz Van Gogha). Chcemy zachowa ksztaty (tre), ale podmieni statystyk kolor贸w (styl).

```python
# Inicjalizacja InstanceNorm
in_layer = nn.InstanceNorm2d(num_features=3)

out_in = in_layer(batch)

print("--- INSTANCE NORM ---")
print("Wynik:\n", out_in[0, 0, :, :])
```

##  Podsumowanie: ciga Architekta

Kiedy czego u偶ywa?

1.  **BatchNorm:**
    *   **Gdzie:** CNN, Computer Vision (ResNet, YOLO).
    *   **Warunek:** Musisz mie du偶y Batch Size (>32). Przy maym Batchu (np. 2) rednia jest zaszumiona i BN psuje trening.
    
2.  **LayerNorm:**
    *   **Gdzie:** NLP, Transformery (BERT, GPT), RNN.
    *   **Zaleta:** Dziaa niezale偶nie od Batch Size. Idealne, gdy dane wejciowe maj r贸偶n dugo.

3.  **InstanceNorm:**
    *   **Gdzie:** Generative AI (GAN, Style Transfer).
    *   **Cel:** Gdy chcesz manipulowa stylem obrazu, a nie jego treci.

4.  **GroupNorm:** (Hybryda)
    *   Dzieli kanay na grupy. U偶ywane w detekcji obiekt贸w, gdy Batch Size musi by may (bo obrazki s wielkie 4K).
