# К NAS: Gdy AI projektuje AI (Neural Architecture Search)

Projektowanie sieci neuronowych to czarna magia. Dlaczego ResNet ma akurat tyle warstw? Bo kto tak sprawdzi i zadziaao.

**NAS (Neural Architecture Search)** automatyzuje ten proces.
Traktujemy architektur sieci jako **Genom**.
*   Genom: `[64, 128, 64]` -> Oznacza sie z 3 warstwami ukrytymi o takich rozmiarach.
*   Genom: `[16]` -> Malutka sie z 1 warstw.

Zastosujemy **Ewolucj**:
1.  **Populacja:** Losujemy 10 r贸偶nych architektur.
2.  **Fitness:** Trenujemy ka偶d przez 1 epok na MNIST. Wynik Accuracy to jej sia.
3.  **Mutacja:** Bierzemy najlepsz sie i losowo zmieniamy liczb neuron贸w lub dodajemy warstw.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import numpy as np
import copy
import random

# Konfiguracja
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
POPULATION_SIZE = 5  # Maa populacja dla szybkoci
GENERATIONS = 3      # Kr贸tka ewolucja
EPOCHS_PER_NET = 1   # Szybki test (tylko 1 epoka)

# Dane (MNIST)
train_loader = DataLoader(
    datasets.MNIST('data', train=True, download=True, transform=transforms.ToTensor()),
    batch_size=64, shuffle=True
)
test_loader = DataLoader(
    datasets.MNIST('data', train=False, transform=transforms.ToTensor()),
    batch_size=1000, shuffle=False
)

print(f"Urzdzenie: {DEVICE}")
```

## Dynamiczna Sie (Budowniczy)

Musimy napisa klas, kt贸ra nie ma sztywnej struktury w `__init__`.
Zamiast tego przyjmuje list `architecture` (np. `[100, 50]`) i dynamicznie buduje warstwy za pomoc `nn.ModuleList`.

```python
class DynamicNet(nn.Module):
    def __init__(self, architecture):
        super().__init__()
        self.architecture = architecture
        self.layers = nn.ModuleList()
        
        # Wejcie: 784 (MNIST 28x28)
        input_dim = 784
        
        # Budujemy warstwy ukryte na podstawie genotypu
        for hidden_dim in architecture:
            self.layers.append(nn.Linear(input_dim, hidden_dim))
            self.layers.append(nn.ReLU())
            input_dim = hidden_dim # Wyjcie tej warstwy to wejcie nastpnej
            
        # Warstwa wyjciowa (zawsze 10 klas)
        self.output_layer = nn.Linear(input_dim, 10)

    def forward(self, x):
        x = x.view(-1, 784) # Spaszczamy obrazek
        for layer in self.layers:
            x = layer(x)
        return self.output_layer(x)

# Test: Stw贸rzmy losow sie
dummy_net = DynamicNet([128, 64]) # 2 warstwy
print("Przykadowa architektura:")
print(dummy_net)
```

## Funkcja Oceny (Fitness)

To funkcja, kt贸ra:
1.  Bierze architektur (list).
2.  Buduje model.
3.  Trenuje go szybko (1 epoka).
4.  Zwraca Accuracy na zbiorze testowym.

To bdzie nasz "koszt 偶ycia" dla danego osobnika.

```python
def evaluate_architecture(arch):
    print(f"К Testuj architektur: {arch} ...", end=" ")
    
    # Budowa modelu
    model = DynamicNet(arch).to(DEVICE)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()
    
    # Szybki trening (1 epoka)
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(DEVICE), target.to(DEVICE)
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
        
        # Przerywamy po 100 batchach dla szybkoci demo (opcjonalne)
        if batch_idx > 100: break 

    # Szybka ewaluacja
    model.eval()
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(DEVICE), target.to(DEVICE)
            output = model(data)
            pred = output.argmax(dim=1)
            correct += pred.eq(target).sum().item()
            
    acc = 100. * correct / len(test_loader.dataset)
    print(f"-> Accuracy: {acc:.2f}%")
    return acc
```

## Ewolucja (Search Loop)

1.  **Inicjalizacja:** Tworzymy losowe "mutanty" (np. `[30]`, `[100, 100]`).
2.  **Mutacja:**
    *   Zmie liczb neuron贸w w warstwie (np. $50 \to 60$).
    *   Dodaj now warstw.
    *   Usu warstw.

Uruchamiamy proces na 3 pokolenia.

```python
# 1. Populacja pocztkowa (losowe listy)
population = [
    [random.randint(32, 128) for _ in range(random.randint(1, 3))]
    for _ in range(POPULATION_SIZE)
]

def mutate(arch):
    """Dokonuje losowej zmiany w architekturze"""
    new_arch = arch.copy()
    choice = random.random()
    
    if choice < 0.33 and len(new_arch) > 1:
        # Usu warstw
        new_arch.pop(random.randint(0, len(new_arch)-1))
    elif choice < 0.66:
        # Zmie rozmiar warstwy
        idx = random.randint(0, len(new_arch)-1)
        new_arch[idx] = int(new_arch[idx] * random.uniform(0.5, 1.5))
    else:
        # Dodaj warstw
        new_arch.insert(random.randint(0, len(new_arch)), random.randint(32, 128))
        
    # Zabezpieczenie przed pust sieci
    if len(new_arch) == 0: return [32]
    return new_arch

# GWNA PTLA NAS
best_arch = None
best_acc = 0

for gen in range(GENERATIONS):
    print(f"\n--- POKOLENIE {gen+1} ---")
    scores = []
    
    # Ocena caej populacji
    for arch in population:
        acc = evaluate_architecture(arch)
        scores.append((acc, arch))
        
        if acc > best_acc:
            best_acc = acc
            best_arch = arch
    
    # Selekcja (Bierzemy 2 najlepsze)
    scores.sort(key=lambda x: x[0], reverse=True)
    survivors = [x[1] for x in scores[:2]]
    
    print(f"Najlepsi w tym pokoleniu: {survivors}")
    
    # Reprodukcja (Mutacja zwycizc贸w)
    new_population = list(survivors) # Elityzm (zostawiamy najlepszych)
    while len(new_population) < POPULATION_SIZE:
        parent = random.choice(survivors)
        child = mutate(parent)
        new_population.append(child)
        
    population = new_population

print("-" * 30)
print(f" ZWYCIZCA NAS: {best_arch}")
print(f"Wynik: {best_acc:.2f}%")
```

```python
# Trenujemy zwycizc "na powa偶nie" (du偶ej)
print("Trenowanie zwyciskiej architektury do pena...")
final_model = DynamicNet(best_arch).to(DEVICE)
# ... tu bymy zrobili normalny trening na 10 epok
print("Gotowe. AI zaprojektowao sobie m贸zg.")
```
